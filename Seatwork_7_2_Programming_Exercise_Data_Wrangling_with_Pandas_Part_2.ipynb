#### 1.) Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). Hint: Be sure to convert the year column to a number as we did in the previous exercise.

```python
meteorites['year'] = meteorites.year.str.slice(6,11) # Update year column to year only
meteorites['year'] = meteorites.year.apply(pd.to_numeric) # convert into numeric
```

```python
meteorites1 = meteorites # to save the changes in year conversion to new variable
meteorites1
```

```python
meteorites1 = meteorites1[(meteorites1['year'] >= 2005) & (meteorites1['year'] <= 2009)] # to filter the meteorites data in range of 2005 to 2009
meteorites1
```

```python
meteorites1 = meteorites1.groupby(['year', 'fall'])['mass (g)'].quantile(0.95) # to create a pivot table and get the 95th percentile
meteorites1
```

```python
from google.colab import files
files = files.upload()
```

```python
taxi = pd.read_csv('2019_Yellow_Taxi_Trip_Data (1).csv')
```

```python
taxi['tpep_dropoff_datetime'] = pd.to_datetime(taxi['tpep_dropoff_datetime']) # to dropoff datetime column to datetime format
```

```python
taxi.set_index("tpep_dropoff_datetime", inplace = True) # to the dropoff datetime as the DataFrame index
```

```python
taxi['Hour'] = taxi.index.hour # to extract the hour from the datetime index and create a new column
```

```python
# Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount, then find the 5 hours with the most tips.
taxi = taxi.groupby('Hour')[['trip_distance', 'fare_amount', 'tolls_amount', 'tip_amount']].sum() # to group the data and get the sum
taxi = taxi["tip_amount"].nlargest(5)
taxi
```
